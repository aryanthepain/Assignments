# Lists
- heterogenous data possible
- duplicate data possible
- order matters
- update and print like array

```python
bicycle=["treak","hi",'bye',"1233", "203", "20.5"]
print(bicycle)

bicycle.append(1233)
print(bicycle)

bicycle.insert(2, "hello")
print(bicycle)

bicycle.pop(3) #last element deleted by default
print(bicycle)

bicycle.remove(1233) #removes the first occurrence of the element
print(bicycle)

bicycle.sort()
print(bicycle)

bicycle.sort(reverse=True)
print(bicycle)

print(sorted(bicycle))
print(bicycle)

print(len(bicycle))
```

# Dictionary

```python
dic={}
print(type(dic))

dic={1:"one",
     2:"two"}
print(dic)

dic[3]="three"
print(dic)

dic.update({4:"four"})
print(dic)

del dic[3]
print(dic)

#key must be immutable
dic[(1,2)]="hi"
print(dic)

print(dic.keys())
print(dic.values())
print(dic.items())

print(dic.get(2))
print(dic.get(5))
```
