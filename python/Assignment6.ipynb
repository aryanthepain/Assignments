## Class Rooms
#### class variables: room_info
#### instance variables: room_number, customer_name
#### class methods: Display_info()
#### instance methods: Check_in(name: string), Check_out() 

```python
class Rooms:
    room_info = []

    def __init__(self):
        self.room_number = 0
        self.customer_name = ""

    def Check_in(self, name):
        occupied_rooms = []
        available_room = []

        #print('\nfrom check in')
        for i in Rooms.room_info:
            #print(i.room_number, i.customer_name)
            occupied_rooms.append(i.room_number)

        #print('\noccupied rooms', occupied_rooms)

        for i in range(1, 26):
            if i not in occupied_rooms:
                available_room.append(i)

        #print('\navailabel rooms', available_room)

        if len(available_room) != 0:
            self.room_number = available_room[0]
            self.customer_name = name
            Rooms.room_info.append(self)

    def Check_out(self):
        for i in Rooms.room_info:
            if i.room_number == self.room_number:
                Rooms.room_info.remove(i)
    
    @classmethod
    def Display_info(cls):
        for i in Rooms.room_info:
            print(i.room_number, i.customer_name)


#print('\nbefore object creation')
#Rooms.Display_info()
    
# customer1 = Rooms()
# Rooms.Display_info()

# customer1.Check_in("Bipul")
# print('\nafter Bipul check in')
# Rooms.Display_info()

# customer2 = Rooms()
# customer2.Check_in("Gautam")
# print('\nafter Gautam check in')
# Rooms.Display_info()

# customer1.Check_out()
# print('\nafter Bipul check out')
# Rooms.Display_info()
```

## Hotel Class

```python
class Hotel(Rooms):
    filepath=Path('Hotel.txt')

    def __init__(self):
        super().__init__()

    @classmethod
    def read(cls):
        # check if file exists
        if not cls.filepath.exists():
            print('Hotel.txt not found, creating a new file')
            cls.filepath.touch()
            return
        
        cls.room_info = []  # Clear previous data
        
        with cls.filepath.open() as f:
            
            while True:
                room_number = f.readline().strip()
                if not room_number:
                    break  # End of file
                
                customer_name = f.readline().strip()
                if not customer_name:
                    break  # Prevents partial reads
                
                obj = Hotel()
                obj.room_number = int(room_number)
                obj.customer_name = customer_name
                cls.room_info.append(obj)

            # sort according to room number
            cls.room_info.sort(key=lambda x: x.room_number)

            print('Rooms loaded from Hotel.txt')
        
        
    @classmethod
    def Display_info(cls):
        Hotel.read()
        for i in Hotel.room_info:
            print(i.room_number, i.customer_name)
    
    @classmethod
    def write_to_file(cls):
        with cls.filepath.open("w") as f:
            for i in cls.room_info:
                f.writelines([str(i.room_number), "\n", i.customer_name, "\n"])

    def Check_out(self):
        Hotel.read()
        for i in Hotel.room_info:
            if i.room_number == self.room_number:
                Hotel.room_info.remove(i)
                break  # Stop loop after removing
        Hotel.write_to_file()

        
    
    def Check_availibiliy(self, type:str):
        type=type.lower()
        
        occupied_rooms = []
        available_room = []

        for i in Hotel.room_info:
            occupied_rooms.append(i.room_number)

        for i in range(1, 26):
            if i not in occupied_rooms:
                available_room.append(i)

        if type=='single':
            start=1
            end=11
        elif type=='double':
            start=11
            end=21
        elif type=='family suite':
            start=21
            end=26
        elif type=='': # no type is given
            start=1
            end=26
        else:
            return -2 #incorrect type

        for i in range(start, end):
            if i in available_room:
                return i
            
        return -1 # unavailable room
    
    def Check_in(self, name:str, type:str=''):
        Hotel.read()
        
        room_number = self.Check_availibiliy(type)

        if room_number == -2:
            print("Wrong room type: use single/double/family suite")
        elif room_number == -1:
            print("No room available of required type")
        else:
            self.room_number = room_number
            self.customer_name = name
            Hotel.room_info.append(self)
            Hotel.write_to_file()


    @classmethod
    def Searchstring(cls, name:str):
        isFirst=True
        name=name.lower()

        for i in Hotel.room_info:
            if name==i.customer_name.lower():
                if isFirst:
                    isFirst=False
                    print('Customer found with details')
                print(i.room_number, i.customer_name)

        if isFirst:
            print('Customer not found')

    @classmethod
    def Search(cls, room_number):
        Hotel.read()

        # go to search string for string input
        if isinstance(room_number, str):
            Hotel.Searchstring(room_number)
            return
        
        for i in Hotel.room_info:
            if room_number == i.room_number:
                print('Customer found with details')
                print(i.room_number, i.customer_name)
                return
                
        print('Customer not found')


```
