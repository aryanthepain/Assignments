# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as mt

np.random.seed(3)
```

## Initialising the problem

```python
x = np.array([2.0,-7.0]) # x values
y = np.array([3.0,9.0])  # y values

eta = 0.01
number_of_iterations = 300
```

```python
# defining weights

# y = m * x + c
m = np.random.normal(scale=1)
m_list = [m]
c = np.random.normal(scale=6)
c_list = [c]

print("Initial m =", m)
print("Initial c =", c)
```

## Implementing gradient descent without using inbuilt function

```python
for i in range(number_of_iterations):
    # calculating predicted values
    y_pred = m*x + c
    
    # calculating gradients
    grad_m = -2 * np.mean(x*(y-y_pred))
    grad_c = -2 * np.mean(y-y_pred)

    # updating values
    m -= eta * grad_m
    c -= eta * grad_c

    # storing values
    m_list.append(m)
    c_list.append(c)

print("m =", m)
print("c =", c)
```

## Visualising the lines

```python
plt.figure()
# all other lines
for i in range(len(m_list)):
    plt.axline((0,c_list[i]), slope=m_list[i], c=(np.random.rand(), np.random.rand(), np.random.rand(), 0.10))
plt.axline((0,c), slope=m, c='k', ls='--') # final line
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Algorithm without using inbuilt function')
plt.show()
```

## Using sklearn

```python
from sklearn.linear_model import SGDRegressor
```

```python
x_shaped = np.reshape(x, (-1,1))
model = SGDRegressor(max_iter=1, eta0=0.01, learning_rate='constant')

m_list=[]
c_list=[]

for i in range(number_of_iterations):
    model.partial_fit(x_shaped,y)
    m=model.coef_[0]
    c=model.intercept_[0]

    # storing values
    m_list.append(m)
    c_list.append(c)

print("m =", m)
print("c =", c)
```

## Visualising

```python
plt.figure()
# all other lines
for i in range(len(m_list)):
    plt.axline((0,c_list[i]), slope=m_list[i], c=(np.random.rand(), np.random.rand(), np.random.rand(), 0.10))
plt.axline((0,c), slope=m, c='k', ls='--') # final line
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Algorithm without using inbuilt function')
plt.show()
```
