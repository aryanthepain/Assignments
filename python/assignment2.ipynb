# Q1

```python
# test cases
tc=[["RED", 2015, 30000, "FERRARI"],
    ["RED", 2005, 40000, "TOYOTA"],
    ["RED", 2008, 60000, "HONDA"],
    ["YELLOW", 2020, 10000, "FERRARI"],
    ["YELLOW", 2022, 5000, "BMW"],
    ]
```

```python
def q1(i):
    # taking input
    color=tc[i][0]
    model=tc[i][1]
    mileage=tc[i][2]
    make=tc[i][3]

    print("Color: ", color)
    print("Model: ", model)
    print("Mileage: ", mileage)
    print("Make: ", make)
    print("Decision to be taken:-")

    # defining tree
    if(color=="RED"):
        if(model>2010):
            print("Buy")
        else:
            if(mileage<50000):
                print("Buy")
            else:
                print("Do NOT Buy")
    else:
        if(color=="YELLOW"):
            if(make=="FERRARI"):
                print("Buy")
            else:
                print("Do NOT Buy")
        else:
            print("Do NOT Buy")

# running the function
for i in range(0,5):
    q1(i)
    print("\n===================\n")

```

# Q2

```python
# defining condition functions

# last level
def hasLowMileage(mileage):
    return mileage<50000
def isFerrari(make):
    return make=="FERRARI"

# second level
def isYellow(color, make):
    if(color=="YELLOW"):
        return isFerrari(make)
    else:
        return False
def isNew(model, mileage):
    if(model>2010):
        return True
    else:
        return hasLowMileage(mileage)

# first level
def checkBuy(color, model, mileage, make):
    if(color=="RED"):
        return isNew(model, mileage)
    else:
        return isYellow(color, make)


# defining the tree
def q2(i):
    # taking input
    color=tc[i][0]
    model=tc[i][1]
    mileage=tc[i][2]
    make=tc[i][3]

    print("Color: ", color)
    print("Model: ", model)
    print("Mileage: ", mileage)
    print("Make: ", make)
    print("Decision to be taken:-")

    # defining tree

    if(checkBuy(color, model, mileage, make)):
        print("BUY")
    else:
        print("Do NOT Buy")

# running the function
for i in range(0,5):
    q2(i)
    print("\n===================\n")
```

```python
def giveIITG(n):
    if(n<1):
        print("Error: give input greater than 1 for optimal results.")
        return
    
    # first line
    print( "I"*(2*n+1) + " " + "I"*(2*n+1) +" "+ "T"*(2*n+1) +' '+ 'G'*(2*n+1) )

    # lines above middle line
    for x in range(n-1):
        print(' '*n+'I'+' '*n +' '+ ' '*n+'I'+' '*n +' '+ ' '*n+'T'+' '*n+' '+'G')

    # middle line
    print(' '*n+'I'+' '*n +' '+ ' '*n+'I'+' '*n +' '+ ' '*n+'T'+' '*n+' '+'G'+ ' '*(n-1)+ 'G'*(n+1))

    # lines below middle line
    for x in range(n-1):
        print(' '*n+'I'+' '*n +' '+ ' '*n+'I'+' '*n +' '+ ' '*n+'T'+' '*n+' '+'G'+ ' '*(2*n-1)+ 'G')

    # last line
    print( "I"*(2*n+1) + " " + "I"*(2*n+1) +" "+ ' '*n+'T'+' '*n +' '+ 'G'*(2*n+1) )


n=int(input("size of IITG?"))
giveIITG(n)
```

```python
def findWords():
    # taking input
    sentence=input("Provide the sentence")

    # creating empty list
    words=[]
    
    # creating empty string
    temp=''
    for i in sentence:
        # end of sentence by full stop
        if(i=='.'):
            break
        # ignoring some of the punctuations
        if(i==' ' or i=="!" or i=="?" or i=="," or i=="(" or i==")"):
            words.append(temp)
            temp=''
            continue

        temp=temp+i
    words.append(temp)
    
    return [sentence,words]

# calling function and printing results
result=findWords()
sentence=result[0]
words=result[1]
print("Sentence: ", sentence)
print("Words:-")
for i in words:
    print(i)
```

```python
def findWordsInDictionary():
    # taking input
    sentence=input("Provide the sentence")

    # creating empty list
    words={}
    # creating empty string
    temp=''
    for i in sentence:
        # end of sentence by full stop
        if(i=='.'):
            break
        # ignoring some of the punctuations
        if(i==' ' or i=='\n' or i=="!" or i=="?" or i=="," or i=="(" or i==")"):
            count=words.get(temp)
            if(not count):
                count=0

            words.update({temp:count+1})
            temp=''
            continue

        temp=temp+i
    count=words.get(temp)
    if(not count):
        count=0
    words.update({temp:count+1})
    
    return (sentence,words)

# calling function and printing results
result=findWordsInDictionary()
sentence=result[0]
words=result[1]
print("Sentence: ", sentence)
print("Words:-")
print(words)

# printing sorted list
print("Sorted List:-")
wordList=list(words.keys())
freqList=list(words.values())

# function to find max index
def findMaxIndex(arr):
    index=0
    max=arr[0]

    for i in range(len(arr)):
        if(arr[i]>max):
            max=arr[i]
            index=i
    
    return index

# printing max frequency word and then setting it to 0
for n in range(len(freqList)):
    maxIndex=findMaxIndex(freqList)
    freqList[maxIndex]=0
    print(wordList[maxIndex])
```
